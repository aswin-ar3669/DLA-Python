```python
import tensorflow as tf

# Create a variable
x = tf.Variable([1, 2, 3], dtype=tf.int32)
print("Variable x:", x)
print("Shape of x:", x.shape)
```

```python
# Initialize a variable
v = tf.Variable(5.0)

# Update the variable using the assign() method
v.assign(10.0)
print("Updated value of v:", v.numpy())

# Increment the variable
v.assign_add(2.0)
print("After incrementing:", v.numpy())

# Decrement the variable
v.assign_sub(1.0)
print("After decrementing:", v.numpy())
```

```python
# Create variables
a = tf.Variable(3.0)
b = tf.Variable(4.0)

# Perform mathematical operations
c = a * b + 2
print("Result of the operation (a * b + 2):", c.numpy())
```

```python
# Create a variable with random initial values
random_var = tf.Variable(tf.random.normal([3, 3], mean=0, stddev=1))
print("Random Variable:")
print(random_var)
```

```python
# Weight and bias initialization for a model
weights = tf.Variable(tf.random.truncated_normal([4, 3], stddev=0.1))
biases = tf.Variable(tf.zeros([3]))

print("Weights:", weights.numpy())
print("Biases:", biases.numpy())
```
