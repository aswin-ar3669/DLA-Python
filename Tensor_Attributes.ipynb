```python
import tensorflow as tf

# Create a tensor
tensor = tf.constant([[1, 2], [3, 4], [5, 6]])

# Get the shape
print("Tensor shape:", tensor.shape)
```

```python
# Create a tensor with a specific data type
tensor = tf.constant([1.0, 2.0, 3.0], dtype=tf.float32)

# Get the data type
print("Data type:", tensor.dtype)
```

```python
# Create a tensor
tensor = tf.constant([[1, 2, 3], [4, 5, 6]])

# Get the rank
rank = tf.rank(tensor)
print("Rank of the tensor:", rank.numpy())
```

```python
# Create a tensor
tensor = tf.constant([[1, 2], [3, 4], [5, 6]])

# Get the size
size = tf.size(tensor)
print("Number of elements in the tensor:", size.numpy())
```

```python
# Create a tensor
tensor = tf.constant([1, 2, 3])

# Check the device placement
print("Device placement:", tensor.device)
```

```python
# Create a tensor
tensor = tf.constant([[1, 2, 3], [4, 5, 6]])

# Get the shape as a list
shape_list = tensor.shape.as_list()
print("Shape as list:", shape_list)
```

```python
# Create a tensor
tensor = tf.constant([1, 2, 3], dtype=tf.int32)

# Convert the data type
tensor_float = tf.cast(tensor, dtype=tf.float32)
print("New data type:", tensor_float.dtype)
```

```python
# Create a trainable tensor
trainable_tensor = tf.Variable([1, 2, 3])

# Check if trainable
print("Is trainable?", trainable_tensor.trainable)
```

```python
# Create a tensor
tensor = tf.constant([1, 2, 3])

# Convert to NumPy array
numpy_array = tensor.numpy()
print("Numpy representation:", numpy_array)
```
